```python
#Check if a given number is prime or not using a for loop
count=0
n= int(input("enter number:"))
for i in range(2,n):
    if n%i==0:
        count+=i
if count==0:
        print("prime",n)
else:
        print("not prime",n)
```

```python
#Print the factorial of a number using a for loop
n= int(input ("enter number: "))
fact=1
for i in range(1,n+1):
    fact=fact*i
print(fact)
```
